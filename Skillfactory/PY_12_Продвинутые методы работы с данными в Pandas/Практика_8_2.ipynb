```python
import pandas as pd
orders = pd.read_csv('data/orders.csv', sep=';')
products = pd.read_csv('data/products.csv', sep=';')
display(orders.head())
display(products.head())
```

```python
orders_products = orders.merge(
    products,
    how='left',
    left_on='ID товара',
    right_on='Product_ID'
)
orders_products.head(10)
```

```python
print(orders.shape)
print(orders_products.shape)
```

```python
mask1 = orders_products['Статус'] == 'Оплачен, формируется к отправке'
mask2 = orders_products['Статус'] == 'Выполнен'
xampl = orders_products[mask1 | mask2]
xampl
```

```python
xampl['result'] = xampl['Количество'] * xampl['Price']
xampl
```

```python
xampl.groupby('ID Покупателя')['result'].sum()
```
